```python
#Function to discriminate between discrete and continuous variables

# float may indicates continuous, but low nÂº of (float or integer) unique values... discrete/categorical

def discrete_continuous (x):
    continuous_df = x.select_dtypes('float64')
    discrete_df = x.select_dtypes('int64')
    
    return continuous_df, discrete_df
```

##### Dealing with customer lifetime value outliers

```python
numerical_df['customer_lifetime_value'].describe()
```

```python
# Checking how many values will be removed if the outliers are removed

iqr = np.nanpercentile(numerical_df['customer_lifetime_value'],75) - np.nanpercentile(numerical_df['customer_lifetime_value'],25)

upper_limit = np.nanpercentile(numerical_df['customer_lifetime_value'],75) + 1.5*iqr
print(upper_limit)

# This checks the number of points that will be removed

extraordinary_points = numerical_df[numerical_df['customer_lifetime_value'] >upper_limit]
display('Rows to be removed: ', len(extraordinary_points))  
```

```python
"""
def detect_outlier(data):
    # find q1 and q3 values
    q1, q3 = np.percentile(sorted(data), [25, 75])
 
    # compute IRQ
    iqr = q3 - q1
 
    # find lower and upper bounds
    lower_bound = q1 - (1.5 * iqr)
    upper_bound = q3 + (1.5 * iqr)
 
    outliers = [x for x in data if x <= lower_bound or x >= upper_bound]
 
    return outliers

detect_outlier(numerical_df['customer_lifetime_value'])
"""
```

```python
# IQR, Upper and Lower Limits of 'customer_lifetime_value'

clv_iqr = np.percentile(numerical_df['customer_lifetime_value'],75) - np.percentile(numerical_df['customer_lifetime_value'],25)
clv_upper_limit = np.percentile(numerical_df['customer_lifetime_value'],75) + 1.5*clv_iqr
clv_lower_limit = np.percentile(numerical_df['customer_lifetime_value'],25) - 1.5*clv_iqr
```

```python
# Remove the outliers from that column.

numerical_df = numerical_df[(numerical_df['customer_lifetime_value']>clv_lower_limit) & (numerical_df['customer_lifetime_value']<clv_upper_limit)].copy()
sns.displot(numerical_df['customer_lifetime_value'])
plt.show()
```

##### Dealing with total claim amount outliers

```python
numerical_df['total_claim_amount'].describe()
```

```python
# Checking how many values will be removed if the outliers are removed

iqr = np.nanpercentile(numerical_df['total_claim_amount'],75) - np.nanpercentile(numerical_df['total_claim_amount'],25)

upper_limit = np.nanpercentile(numerical_df['total_claim_amount'],75) + 1.5*iqr
print(upper_limit)

# This checks the number of points that will be removed

extraordinary_points = numerical_df[numerical_df['total_claim_amount'] >upper_limit]
display('Rows to be removed: ', len(extraordinary_points))  
```

```python
# IQR, Upper and Lower Limits of 'total claim amount'

tca_iqr = np.percentile(numerical_df['total_claim_amount'],75) - np.percentile(numerical_df['total_claim_amount'],25)
tca_upper_limit = np.percentile(numerical_df['total_claim_amount'],75) + 1.5*tca_iqr
tca_lower_limit = np.percentile(numerical_df['total_claim_amount'],25) - 1.5*tca_iqr
```

```python
# Remove the outliers from that column.

numerical_df = numerical_df[(numerical_df['total_claim_amount']>tca_lower_limit) & (numerical_df['total_claim_amount']<tca_upper_limit)].copy()
sns.displot(numerical_df['total_claim_amount'])
plt.show()
```
